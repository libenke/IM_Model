## Evolution of stretch ratio $\lambda$
definition:
$$\lambda = L/L_0$$
here, $L$ is the current tube length, $L_0$ is the equilibrium value.  
The evolution of $\lambda$ is (ref: Pearson, et. al. *Journal of Rheology* __1989__, 33, 517-535.) :  
$$
\frac{d\lambda}{dt} = \lambda\boldsymbol{\kappa}:\mathbf{S} - \frac{\lambda-1}{\tau_R}
$$
With extending to finite extensibility (FENE), $\lambda$ was replaced by $F(\lambda)$  
1. (ref: Ianniruberto, G.; Marrucci, G. *Journal of Rheology* __2001__, 45, 1305-1318.):   
$$
F(\lambda)=\frac{\lambda_{max}-1}{\lambda_{max}-\lambda}\lambda
$$
here $\lambda_{max} = L_{max}/L_0$ is the maximum stretch ratio,  
or (ref: Costanzo, S. et. al. *Macromolecules* __2016__, 49, 3925-3935.):  
$$\lambda_{max} = a/b$$
here, $a$ is the tube diameter, $b$ is the Kuhn length.  
Thus, the evolution of $\lambda$ during shear become (ref: Ianniruberto, G.; Marrucci, G. *Journal of Rheology* __2001__, 45, 1305-1318.):  
$$
\begin{align}
\frac{d\lambda}{dt} &= \lambda\boldsymbol{\kappa}:\mathbf{S} - \frac{F(\lambda)-1}{\tau_R} \\
                    &= \lambda\boldsymbol{\kappa}:\mathbf{S} - \frac{\lambda_{max}}{\tau_R}\frac{\lambda-1}{\lambda_{max}-\lambda}
\end{align}$$
2. ref: Costanzo, S. et. al. *Macromolecules* __2016__, 49, 3925-3935.  
The expression comes from the Langevin dependence 
o 
elongation on force (see section 2.2.2 oColby and Rubinstein, *f Polymer Physi*c__2003__eines 75-7  7).
$$\begin{align}
F(\lambda)&=\frac{\mathcal{L}^{-1}(\lambda/\lambda_{max})}{3\lambda/\lambda_{max}} \\
&=\frac{\lambda_{max}^2-\frac{\lambda^2}{3}}{\lambda_{max}^2-\lambda^2}
\end{align}$$
here, $\mathcal{L}^{-1}$ is the inverse langevin function, and the langevin function is $y = \mathcal{L}(x) = \coth{x}-1/x$.  
The most common used approximation of the inverse langevin function is from Cohen:  
ref: Jedynak, R. Approximation of the inverse Langevin function revisited. *Rheologica Acta* __2015__, 54, 29-39.  
ref: Darabi, E.; Itskov, M. A simple and accurate approximation of the inverse Langevin function. *Rheologica Acta* __2015__, 54, 455-459.
$$
\mathcal{L}^{-1}(y) = y\frac{3-y^2}{1-y^2}
$$
For this approximation, $F(0)$ is not equal to 1, so, it is commonly normalized by $F(0)$ as:  
ref: chapter 11.3.2 of Dealy, J. M. Larson, R. G. Read, D. J. "Structure and rheology of molten polymers, from structure to flow behavior and back again." __2018__, Carl Hanser Verlag GmbH Co KG.  
ref: Taghipour, H.; Costanzo, S.; Vlassopoulos, D.; van Ruymbeke, E.; Hawke, L. G. D. Entangled linear polymers in fast shear flows: Comparison of tube-model predictions and experimental data. *Journal of Rheology* __2021__, 65, 1111-1137.  
$$
F(\lambda)=\left(\frac{\lambda_{max}^2-\frac{\lambda^2}{3}}{\lambda_{max}^2-\lambda^2}\right)\left(\frac{\lambda_{max}^2-1}{\lambda_{max}^2-\frac{1}{3}}\right)
$$

## Finger Tensor $\mathbf{B}$ and Cauchy Tensor $\mathbf{C}$
Ref: Dealy, J. M. Larson, R. G. Read, D. J. "Structure and rheology of molten polymers, from structure to flow behavior and back again." __2018__, Carl Hanser Verlag GmbH Co KG.  
Displacement Gradient Tensor $F_{ij}$:  
$$
F_{ij}(t_0,t_1) = \frac{\partial x_i(t_1)}{\partial x_j(t_0)}
$$
Inverse of Displacement Gradient Tensor $E_{ij}$:  
$$
E_{ij}(t_0,t_1) = F_{ij}^{-1}(t_0,t_1) = \frac{\partial x_i(t_0)}{\partial x_j(t_1)}
$$
#### For simple shear:  
$$\mathbf{F} (t_0,t_1)=\begin{pmatrix}
 1 & \gamma (t_1)-\gamma (t_0) & 0\\
 0 & 1 & 0\\
 0 & 0 &1
\end{pmatrix}$$
$$\mathbf{E} (t_0,t_1)=\begin{pmatrix}
 1 & \gamma (t_0)-\gamma (t_1) & 0\\
 0 & 1 & 0\\
 0 & 0 &1
\end{pmatrix}$$  
The Cauchy Tensor, also known as right Cauchy–Green deformation tensor, $\mathbf{C}$:  
$$
\begin{align}
\mathbf{C} &= \mathbf{F}^T\cdot\mathbf{F} \\
&=\begin{pmatrix}
 1 & \gamma & 0\\
 \gamma & 1+\gamma^2 & 0\\
 0 & 0 &1
\end{pmatrix}
\end{align}
$$
The Finger Tensor, also known as left Cauchy–Green deformation tensor, $\mathbf{B}$:  
$$\begin{align}
\mathbf{B} &= \mathbf{F}\cdot\mathbf{F}^{T} \\
&=\begin{pmatrix}
 1+\gamma^2 & \gamma & 0\\
 \gamma & 1 & 0\\
 0 & 0 &1
\end{pmatrix}
\end{align}$$
Sometimes, Finger strain tensor is also defined by inverse of Cauchy tensor as (ref: Milner, S. T. *Journal of Rheology* __2001__, 45, 1023-1028.):  
$$
\mathbf{B} = \mathbf{C}^{-1} = \mathbf{E}\cdot\mathbf{E}^T = \begin{pmatrix}
 1+\gamma^2 & -\gamma & 0\\
 -\gamma & 1 & 0\\
 0 & 0 &1
\end{pmatrix}
$$
Here, the diagonal is the same, while the $B_{0,1}, B_{0,2}, B_{1,2}$ parts are negative.  

#### For rigid body rotation:
$$\mathbf{F} (t_0,t_1)=\begin{pmatrix}
 \cos \theta & \sin \theta & 0\\
 -\sin \theta & \cos \theta & 0\\
 0 & 0 &1
\end{pmatrix}$$
$$\mathbf{B} = \mathbf{C} = \begin{pmatrix}
 1 & \gamma & 0\\
 \gamma & 1 & 0\\
 0 & 0 &1
\end{pmatrix}$$

## Doi-Edwards Nonlinear Strain Measure $\mathbf{Q}\left[\mathbf{E}(t,t')\right]$
in classic DE Model, the orientation tensor $\mathbf{S}$ could be expressed as:  
$$
\mathbf{S} = \int_{-\infty}^t \frac{dt'}{\tau_d} \exp\left(-\frac{t-t'}{\tau_d}\right) \mathbf{Q}[\mathbf{E}(t,t')]
$$
DEMG model add a scalar $\lambda$ describing stretch:  
$$
\dot{\lambda} = \lambda\boldsymbol{\kappa}:\mathbf{S} - \frac{1}{\tau_s}(\lambda-1)
$$
Last the stress is:  
$$
\boldsymbol{\sigma} = 5G_N^0\lambda^2\mathbf{S}
$$
Tensor $\mathbf{Q}\left[\mathbf{E}(t,t')\right]$ is the deformation-dependent "universal" tensor of Doi-Edwards defined by the deformation gradient history $\mathbf{E}(t,t')$:  
$$
\mathbf{Q} \equiv \left \langle \frac{\mathbf{E\cdot u'E\cdot u'}}{\left | \mathbf{E\cdot u'} \right | ^2} \right \rangle _0
$$
where $\mathbf{u}'$ is a unit vector distributed randomly over the unit sphere and $<>$ represents an ensemble average over all $\mathbf{u}'$. The tensor $\mathbf{Q}$ is derived using the "independent alignment approximation" (IAA), ref: Doi, M. A constitutive equation derived from the model of doi and edwards for concentrated polymer solutions and polymer melts. *Journal of Polymer Science: Polymer Physics Edition* __1980__, 18, 2055-2067.

## Approximation of nonlinear strain measure $\mathbf{Q}\left[\mathbf{E}(t,t')\right]$
1. A highly accurate approximation for this strain measure is:  
ref: chapter 11.3 of Dealy, J. M. Larson, R. G. Read, D. J. "Structure and rheology of molten polymers, from structure to flow behavior and back again." __2018__, Carl Hanser Verlag GmbH Co KG.
$$\mathbf{Q} \approx  \left(\frac{5}{J-1}\right)\mathbf{B} - \left[\frac{5}{(J-1)(I_2+13/4)^{1/2}}\right]\mathbf{C}$$
here:  
$$J\equiv I_1+2(I_2+13/4)^{1/2}$$
$$I_1\equiv \mathrm{Tr}(\mathbf{B})$$
$$I_2\equiv \mathrm{Tr}(\mathbf{C})$$
$\mathbf{B}$ is the Finger tensor, and $\mathbf{C}$ is the Cauchy tensor.  
2. Marrucci et al. proposed another approximation of this strain measure, which gives a much improved prediction of the normal stress ratio in shear, namely $-N_2/N_1=1/4$ in the limit of small strains, as compared to Doi-Edwards value of 1/7. ref: Milner, S. T. Improved model of nonaffine strain measure. *Journal of Rheology* __2001__, 45, 1023-1028.
$$
\mathbf{Q}(\mathbf{E})=\frac{\mathbf{C}^{-1/2}}{\mathrm{Tr}(\mathbf{C^{-1/2}})}
$$
here:
$\mathbf{C}^{-1}=\mathbf{E}\cdot\mathbf{E}^T$ is the finger tensor. $\mathbf{E}$ is the inverse of displacement gradient tensor.


## The Classical CCR Model
ref: Ianniruberto, G.; Marrucci, G. On compatibility of the Cox-Merz rule with the model of Doi and Edwards. *Journal of Non-Newtonian Fluid Mechanics* __1996__, 65, 241-246.  
Classical CCR Model use $r$ to describe the tube loss rate (constraint release rate) due to convective flow:  
$r$: tube loss rate  
$$\begin{align}
r &= \boldsymbol{\kappa}:\mathbf{S} - \frac{1}{L}\frac{dL}{dt}
\end{align}$$
here, $L=\lambda L_0$ is the average tube length. and $r = \boldsymbol{\kappa}:\mathbf{S}$ at steady shear flow.  
$$\frac{d\lambda}{dt} = \lambda\boldsymbol{\kappa}:\mathbf{S} - \frac{\lambda F(\lambda)-1}{\tau_R}$$
$$F(\lambda)=\left(\frac{\lambda_{max}^2-\frac{\lambda^2}{3}}{\lambda_{max}^2-\lambda^2}\right)\left(\frac{\lambda_{max}^2-1}{\lambda_{max}^2-\frac{1}{3}}\right)$$
Define an effective time $\tau_{eff}$ to replace the disengagement time $\tau_d$ in DE model:  
$$
\frac{1}{\tau_{eff}} = \frac{1}{\tau_d} + \beta r
$$
Here $\beta$ is an unknown coefficient, here for new CCR model, $\beta = 0.18$.  
The classical DE model of $\mathbf{S}(t)$ is replaced to be:  
$$
\mathbf{S}(t) = \int_{-\infty}^t \left[\frac{1}{\tau_{eff}(t')}\right] \exp \left[-\int_{t'}^t \frac{dt''}{\tau_{eff}(t'')}\right] \mathbf{Q}\left[\mathbf{E}(t,t')\right] dt'
$$

## Modified CCR Model with Loss of Entanglement
Add the loss of entanglement effect to the modeling.  
Ref: Ianniruberto, G.; Marrucci, G. Convective constraint release (CCR) revisited. *Journal of Rheology* __2014__, 58, 89-102.  
The loss rate of entanglements due to convection is:  
$$
r_{loss} = \beta r \nu = \beta\left(\boldsymbol{\kappa}:\mathbf{S} - \frac{1}{L}\frac{dL}{dt}\right)\nu
$$
Here, $\nu$ is the fraction of number of entanglments at any time $t$.  
The rate of entanglement re-creation rate $r_{recr}$ is:  
$$
r_{recr} = \frac{1-\nu}{\tau_d}
$$
Thus,
$$
\frac{d\nu}{dt} = r_{loss}-r_{recr} = -\beta\left(\boldsymbol{\kappa}:\mathbf{S} - \frac{1}{L}\frac{dL}{dt}\right)\nu + \frac{1-\nu}{\tau_d}
$$

The diameter $a$, tube length $L$, disengagement time $\tau_d$:  
$$
\begin{align}
a = a_{\mathrm{eq}} \nu^{-1/2} \\
L=\lambda L_{\mathrm{eq}} \nu^{1/2} \\
\tau_d = \tau_{\mathrm{d,eq}} \nu^{1.2}
\end{align}$$
